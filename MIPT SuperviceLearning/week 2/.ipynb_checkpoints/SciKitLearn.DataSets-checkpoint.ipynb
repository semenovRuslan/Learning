модуль предназначен для генерации модельных данных с задаными параметрами которые удобно отображать на графиках. Функции для генераци:

1. make_circles
позволяет сгенерировать dataset выглядящая на плоскости как 2  вложенные окружности

```python
circles=ds.make_circles()
```

```python
print 'features: {}'.format(circles[0][:10])
print "targets: {}", format(circles[1][:10])
```

```python
#визуализируем это в график:
from matplotlib.colors import ListedColormap

colors = ListedColormap(['red', 'yellow'])

pyplot.figure(figsize(8,8))
pyplot.scatter(map(lambda x:x[0], circles[0]), map(lambda x:x[1], circles[0]), c=circles[1], cmap=colors)
# 1 - pyplot.scatter(0.63742399, 0.77051324, cmap='black')
# 0 - pyplot.scatter(0.79369176,  0.10026659, cmap='black')
```

```python
#make function
def plot_2D_dataset(data, colors):
    pyplot.figure(figsize(8,8))
    pyplot.scatter(map(lambda x:x[0], data[0]), map(lambda x:x[1], data[0]), c=data[1], cmap=colors)
```

```python
noisy_circles=ds.make_circles(noise=0.05)
```

```python
plot_2D_dataset(noisy_circles, colors)
```

2. make_classification()
позволяет гибко создавать наборы данных которые нас интересуют. В частности мы можем задать:
- кол-во объектов которые нас интересуют
- кол-во признаков
- сколько признаков информативные и сколько избыточные
- можно ввести повторяющиеся признаки

```python
simple_classification_problem=ds.make_classification (n_features=2, n_informative=1,
                                                     n_redundant=1, n_clusters_per_class=1,
                                                     random_state=1)
```

```python
plot_2D_dataset(simple_classification_problem, colors)
```

```python
#усложним 4 класса и 2 информативных признака
classification_problem = ds.make_classification (n_features=2, n_informative=2, n_classes=4,
                                                 n_redundant=0, n_clusters_per_class=1,
                                                 random_state=1)
plot_2D_dataset(classification_problem, ListedColormap(['red', 'yellow', 'green', 'blue']))
```

визуализируем наши данные на графике для этого перенесем данные в таблицу

```python
from pandas import DataFrame
```

```python
iris_frame=DataFrame(iris["data"])
iris_frame.columns=iris["feature_names"] #rename the columns
iris_frame["target"]=iris["target"]#add result property
iris_frame.head(5)
```

```python
#replace label of class to name of classes:
iris_frame.target=iris_frame.target.apply(lambda x: iris["target_names"][x])
iris_frame.head(5)
```

```python
# выведем гистограмму распределения свойства 'sepal length (cm)' в классе 'setosa' 
iris_frame[iris_frame.target=="setosa"].hist('sepal length (cm)')
```

```python
pyplot.figure(figsize(20,24))

plot_number=0
for feature_name in iris['feature_names']:
    for target_name in iris["target_names"]:
        plot_number+=1
        pyplot.subplot(3,4,plot_number)
        pyplot.hist(iris_frame[iris_frame.target==target_name][feature_name])
        pyplot.title(target_name)
        pyplot.xlabel("cm")
        pyplot.ylabel(feature_name[:-4])
    
```

```python
# или быстрее
import seaborn as sns
sns.pairplot(iris_frame, hue = 'target')
?sns.set()
sns.set(font_scale = 1.3)
data = sns.load_dataset("iris")
sns.pairplot(data, hue = "species")
```
